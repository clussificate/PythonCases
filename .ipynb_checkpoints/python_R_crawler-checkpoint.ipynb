```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
def namefunc(movie):
    names = [x.findChild('span',attrs = {'class':'title'}).string for x in movie]
    return names
def scorefunc(movie):
    scores = [float(str(x.findChild('span',attrs={'rating_num'}).string)) for x in movie]
    return scores
def numfunc(movie):
    num = [x.findChild('div',attrs={'class':'star'}).text.split('\n')[4] for x in movie]
    num = [int(str(re.sub('\D', '', x))) for x in num]
    return num
def getinfo(url):
    r = requests.get(url)
    r.encoding = 'utf-8'
    soup_packtpage = BeautifulSoup(r.text,'html.parser')
    movie = soup_packtpage.findAll('div',attrs={'class':'info'})
    names = namefunc(movie)
    scores = scorefunc(movie)
    num = numfunc(movie)
    res = {'names': names, 'scores': scores, 'num': num}
    return res
```

```python
urls = []
index = range(0,250,25)
for x in index:
    urls.append("http://movie.douban.com/top250?start="+str(x)+'&filter=&type=')
urls
```

```python
res = {'names':[],'scores':[],'num':[]}
for url in urls:
    new = getinfo(url)
    res['names'].extend(new['names'])
    res['scores'].extend(new['scores'])
    res['num'].extend(new['num'])
```

```python
import pandas as pd
df = pd.DataFrame(res)
df.head()
```
