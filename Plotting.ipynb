```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn import datasets
from sklearn import linear_model
```

```python
boston = datasets.load_boston()
yb = boston.target.reshape(-1, 1)
Xb = boston['data'][:,5].reshape(-1, 1)
```

```python
plt.ylabel('value of house /1000 ($)')
plt.xlabel('number of rooms')
# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit( Xb, yb)
# Plot outputs
plt.scatter(Xb, yb,  color='black')
plt.plot(Xb, regr.predict(Xb), color='blue',
         linewidth=3)
```

```python
# 直方图
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
 
# 数据的直方图
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
 
 
plt.xlabel('Smarts')
plt.ylabel('Probability')
#添加标题
plt.title('Histogram of IQ')
#添加文字
plt.text(60, .025, r'$mu=100, sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
iris = sns.load_dataset("iris")
```

```python
iris_1 = iris[iris['species']=='virginica'].loc[:,['sepal_length','sepal_width']]
iris_2 = iris[iris['species']=='setosa'].loc[:,['sepal_length','sepal_width']]
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('white')
fig,ax = plt.subplots()
plt.scatter(iris_1.sepal_length,iris_1.sepal_width,color = 'red',marker='o')
plt.scatter(iris_2.sepal_length,iris_2.sepal_width,color = 'blue',marker='*')
```
